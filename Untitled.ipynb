```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.2, random_state=1)

def create_model(learning_rate=0.001, hidden_units=11):
    model1 = Sequential()
    model1.add(Dense(hidden_units, activation='sigmoid', input_dim=10))
    model1.add(Dense(1, activation='sigmoid'))
    optimizer = Adam(learning_rate=learning_rate)
    model1.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model1

learning_rates = [0.1, 0.001, 0.0001]
hidden_units_list = [11, 20, 30]
batch_sizes = [16, 32, 64]
epochs_list = [50, 100, 200]

for lr in learning_rates:
    for hidden_units in hidden_units_list:
        for bs in batch_sizes:
            for epochs in epochs_list:
                model1 = create_model(learning_rate=lr, hidden_units=hidden_units)
                hist = model1.fit(Xtrain, Ytrain, batch_size=bs, epochs=epochs, verbose=0, validation_data=(Xtest, Ytest))
                
                # Evaluate the model on the test set
                val_loss, val_accuracy = model1.evaluate(Xtest, Ytest, verbose=0)
                
                # Print or store the results
                print(f'LR={lr}, Hidden Units={hidden_units}, Batch Size={bs}, Epochs={epochs}: Val Loss={val_loss}, Val Accuracy={val_accuracy}')

                # Optionally, plot the training history
                plt.plot(hist.history['loss'], label='Training Loss')
                plt.plot(hist.history['val_loss'], label='Validation Loss')
                plt.legend()
                plt.show()
```

```python
y_pred1 = model1.predict(Xtest)
y_pred1 = y_pred1.argmax(axis=1)
accuracy_score(Ytest,y_pred1)
```
